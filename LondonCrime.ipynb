```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt # plot charts

colors = ['#105B63', '#FFFAD5','#FFD34E','#DB9E36','#BD4932']

import os
print(os.listdir("./"))
```

```python
df = pd.read_csv('./LondonCrime24months.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['MinorText'].unique()
```

```python
weaponData = df.loc[df['MinorText'] == 'Possession of Article with Blade or Point']
```

```python
weaponData.sum()
```

```python
crimeByMonth = weaponData.iloc[:,3:]
crimeByMonth.head()
```

```python
plt.rcParams['figure.figsize'] = [20, 5]
plt.plot(crimeByMonth.sum())
plt.ylabel('Number of knife crimes committed')
plt.xlabel('Month & Year')
plt.show()
```

```python
histogram_intersection = lambda a, b: np.minimum(a, b).sum().round(decimals=1)
df = pd.DataFrame(crimeByMonth,
                  columns=)
df.corr(method=histogram_intersection)
```

```python
explode = (0.1,) + tuple([0 for i in range(9)])
plt.figure(figsize=(20,20))
plt.pie(crimeByMonth, explode = explode, labels=None, autopct='%1.1f%%', shadow=True, startangle= 90)
plt.legend(loc=3)
plt.show()
```
